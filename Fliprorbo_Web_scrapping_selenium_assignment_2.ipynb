```python
import pandas as pd
import selenium
from selenium import webdriver
```

```python
driver=webdriver.Edge("msedgedriver.exe")
```

```python
url= "https://www.naukri.com/"
driver.get(url)
```

```python
#finding element for job search bar
search_job=driver.find_element_by_id("qsb-keyword-sugg")
```

```python
# writing on search bar
search_job.send_keys("Data Analyst")
```

```python
#finding element for job location bar
search_loc=driver.find_element_by_id("qsb-location-sugg")
```

```python
# writing on search bar
search_loc.send_keys("Bangalore")
```

```python
#do click using xpath function
search_btn=driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
#Now we will see in the window opened by webdriver whether the webpage has opened or not.Lets check it.
#Now first craete 4 empty list.In this list we will store data which will be scrapped.
#We have craeted 4 empty lists for 4 features which we have to extract
#job title ,company names ,locations_list,experience_list
```

```python
title_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
title_tags
```

```python
#now we will extarct text from web element
job_titles=[]
for i in title_tags:
    job_titles.append(i.text)
job_titles
```

```python
companies_tag=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tag
```

```python
company_names=[]
for i in companies_tag:
    company_names.append(i.text)
company_names    
```

```python
##here li is the parent tag and span is the tag containing the data therefore used at end
#Also we can span[1] or span[2] if two span tag present..
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span")
location_tags
```

```python
location_list=[]
for i in location_tags:
    location_list.append(i.text)
location_list    
```

```python
##here li is the parent tag and span is the tag containing the data therefore used at end
experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']//span")
experience_tags
```

```python
experience_list=[]
for i in experience_tags:
    experience_list.append(i.text)
experience_list    
```

```python
jobs=pd.DataFrame()
jobs["title"]=job_titles[:10]
jobs["company"]=company_names[:10]
jobs["experience_required"]=experience_list[:10]
jobs["location"]=location_list[:10]
```

```python
jobs
```
