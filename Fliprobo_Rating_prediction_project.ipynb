```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df_final.drop(columns=["Model"],axis=1,inplace=True)
```

```python
#new column which give idea about review length
df_final["length"]=df_final.Review.str.len()
df_final.head(5)
```

```python
#Label coding  5,4,3,2 and 1
df_final["rating"].replace({"5":5,"4":4,"3":3,"2":2,"1":1},inplace=True)
```

```python
#Convert all messages into lower case
df_final["Review"]=df_final["Review"].str.lower()
df_final.head()
```

```python
#Regular Expressions
# Replace email addresses with "email"
df_final["Review"]=df_final["Review"].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$','emailaddress')

#Replace URLs with "webaddress"
df_final["Review"]=df_final["Review"].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$',"webaddress")

#Replace money symbols with "moneysymb" ( can be typed with Alt key + 156)
df_final["Review"]=df_final["Review"].str.replace(r'|\$',"dollars")

#Replace 10 digit phone numbers(formats include paranthesis,spaces,no spaces,dashes) with "phonenumber"
df_final["Review"]=df_final["Review"].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$','phonenumber')

#Replace numbers with 'numbr'
df_final["Review"]=df_final["Review"].str.replace(r'\d+(\.\d+)?',"numbr")

#Remove punctuation
df_final["Review"]=df_final["Review"].str.replace(r'[^\w\d\s]',' ')
```

```python
#replace white space between terms with a single space
df_final["Review"]=df_final["Review"].str.replace(r'\s+',' ')

#Remove Leading and Trailing white space
df_final["Review"]=df_final["Review"].str.replace(r'^\s+|\s?$', '')
```

```python
df_final.head()
```

```python
#Remove stopwords
import string
import nltk
from nltk.corpus import stopwords

stop_words=set(stopwords.words("english")+['u','u','ur','4','2','im','dont','doin','ure'])
```

```python
df_final["Review"]=df_final["Review"].apply(lambda x:' '.join(term for term in x.split() if term not in stop_words))
```

```python
#new column clean length after punctuation,stopwords removal
df_final["clean_length"]=df_final.Review.str.len()
df_final.head()
```

```python
#total length removal
print("original length",df_final.length.sum())
print("Clean length",df_final.clean_length.sum())
```

```python
#convert text into vector using TF-IDF
#instantiate MultinomialNB classifier
#split feature and label

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.tree import DecisionTreeClassifier

tf_vec=TfidfVectorizer()

naive=MultinomialNB()

features=tf_vec.fit_transform(df_final["Review"])

x=features
y=df_final["rating"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=42)

naive.fit(xtrain,ytrain)

y_pred=naive.predict(xtest)
```

```python
print("final score =",accuracy_score(ytest,y_pred))
```

```python
print(classification_report(ytest,y_pred))
```

```python
conf_mat=confusion_matrix(ytest,y_pred)
conf_mat
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()

dtc.fit(xtrain,ytrain)

y_pred=dtc.predict(xtest)

print("final score =",accuracy_score(ytest,y_pred))
print(classification_report(ytest,y_pred))
print(confusion_matrix(ytest,y_pred))
```

```python
from sklearn.svm import SVC
svc=SVC()

svc.fit(xtrain,ytrain)

y_pred=svc.predict(xtest)

print("final score =",accuracy_score(ytest,y_pred))
print(classification_report(ytest,y_pred))
print(confusion_matrix(ytest,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier()

RFC.fit(xtrain,ytrain)

y_pred=RFC.predict(xtest)

print("final score =",accuracy_score(ytest,y_pred))
print(classification_report(ytest,y_pred))
print(confusion_matrix(ytest,y_pred))
```
