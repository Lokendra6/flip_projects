```python
#Q2 Scrape the details team India’s international fixtures from bcci.tv.
#Url = https://www.bcci.tv/.
#You need to find following details: Match title (I.e. 1st ODI), Series, Place, Date, Time
```

```python
driver=webdriver.Edge("msedgedriver.exe")
```

```python
driver.get("https://www.bcci.tv/.")
```

```python
from selenium.webdriver.common.action_chains import ActionChains

option=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']")
ActionChains(driver).move_to_element(option).perform()
driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").click()

```

```python
title=[]
series=[]
place=[]



for t in driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']"):
    title.append(t.text)
for s in driver.find_elements_by_xpath("//div[@class='fixture__teams']"):
    series.append(s.text.replace('\n',' '))
for p in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    place.append(p.text)
```

```python
date=[]
time=[]
for d in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']"):
    date.append(' '.join(d.text.replace('\n',' ').split(' ',3)[:-1]))
    time.append(d.text.replace('\n',' ').split(' ',3)[-1])
```

```python
df = pd.DataFrame(list(zip(title,series,place,date,time)),columns=['title','series','place','date','time'])
df.head()
```

```python
col_name=[]
des = []
expn=[]

for n,elements in enumerate(driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/*/td")):

    if n%2 == 0:
        if n == 0:
            col_name.append(elements.text)
        else:
            expn.append(elements.text)

    if n%2 == 1:
        if n == 1:
            col_name.append(elements.text)
        else:
            des.append(elements.text)

df = pd.DataFrame(list(zip(expn,des)),columns=col_name)
```

```python
#Q4:Scrape the details of State-wise GDP of India from statisticstime.com.Url = http://statisticstimes.com/
#You have to find following details:A) Rank B) State C) GSDP at current price (19-20) D) GSDP at current price (18-19) 
#E) Share(18-19)F) GDP($ billion)
#Note: - From statisticstimes home page you have to reach to economy page through code.
```

```python
driver=webdriver.Edge("msedgedriver.exe")
driver.get("http://statisticstimes.com/")

tab_hover = driver.find_elements_by_xpath("//button[@class='dropbtn']")[1]

ActionChains(driver).move_to_element(tab_hover).perform()
    
time.sleep(2)

driver.find_elements_by_xpath("//div[@class='dropdown-content']")[1].click()
    
driver.find_element_by_xpath("//ul[@style='list-style-type:none;margin-left:20px;']/*/a").click()

rows = driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr")

records = []
for row in rows:
    temp = ''
    for elements in row.find_elements_by_xpath(".//td"):
        temp += elements.text.replace(',','') + ','
    temp = temp[:-1]
    records.append(temp.split(','))

df = pd.DataFrame(records,columns=['Rank','State','GSDP(18-19)','GSDP(17-18)','Share','GPD(billion)','.','.']).drop('.',axis=1)
```

```python
#Q5. Scrape the details of trending repositories on Github.com.Url = https://github.com/You have to find the following details:
#A) Repository title B) Repository descriptionC) Contributors countD) Language used
```

```python

driver=webdriver.Edge("msedgedriver.exe")
    
driver.get("https://github.com/")
    
time.sleep(2)
title_l,des_l,con_count_l,lang_l = [],[],[],[]

driver.find_elements_by_xpath("//summary[@class='HeaderMenu-summary HeaderMenu-link px-0 py-3 border-0 no-wrap d-block d-lg-inline-block']")[1].click()
        
driver.find_element_by_xpath("//a[contains(@data-ga-click,'Trending')]").click()
    
    
for block in driver.find_elements_by_xpath("//article[@class='Box-row']"):
    title_l.append(block.find_element_by_xpath(".//h1[@class='h3 lh-condensed']").text)
    des_l.append(block.find_element_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']").text)
    con_count_l.append(block.find_elements_by_xpath(".//a[@class='Link--muted d-inline-block mr-3']")[1].text)
    try:
        lang_l.append(block.find_element_by_xpath(".//span[@itemprop='programmingLanguage']").text)
    except NoSuchElementException:
        lang_l.append('NA')    
        
df = pd.DataFrame(list(zip(title_l,des_l,con_count_l,lang_l)),columns=['title','description','countributors_count','language'])
    
df

```

```python
#Q6. Scrape the details of top 100 songs on billboard.com.Url = https://www.billboard.com/
#You have to find the following details:A) Song nameB) Artist nameC) Last week rankD) Peak rankE) Weeks on board
#Note: - From the home page you have to click on the charts option then hot 100-page link through code.
```

```python
driver=webdriver.Edge("msedgedriver.exe")
    
driver.get("https://www.billboard.com/")

driver.find_elements_by_xpath("//a[@class='header__subnav__link text--uppercase font--semi-bold']")[2].click()

song_l,artist_l,last_week_l,peak_l,wks_on_chrts_l = [],[],[],[],[]
    
for song in driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']"):
    song_l.append(song.text)
        
for artist in driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']"):
    artist_l.append(artist.text)
        
for last_week in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']"):
    last_week_l.append(last_week.text)
        
for peak in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']"):
    peak_l.append(peak.text)
        
for wks_on_chrt in driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']"):
    wks_on_chrts_l.append(wks_on_chrt.text)
        
    
df = pd.DataFrame(list(zip(song_l,artist_l,last_week_l,peak_l,wks_on_chrts_l)),columns=['song_title','artist','last_week','peak','weeks_on_charts'])
df      
```

```python
#Q7. Scrape the details of Data science recruiters from naukri.com.Url = https://www.naukri.com/
#You have to find the following details:A) NameB) DesignationC) CompanyD) Skills they hire forE) Location
#Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code
```

```python
driver=webdriver.Edge("msedgedriver.exe")
    
driver.get("https://www.naukri.com/")

driver.find_element_by_id('qsb-keyword-sugg').send_keys('Data Science')

driver.find_element_by_class_name('btn').click()
    
time.sleep(2)
    
# scraping the data:

job_titles, job_company = [],[]

job_location,job_skills  = [],[]

for title in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    job_titles.append(title.text)

for company_name in driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']"):
    job_company.append(company_name.text)
        
for skills in driver.find_elements_by_xpath("//ul[@class='tags has-description']"):
    temp = skills.text
    job_skills.append(temp.replace('\n','|'))     
        
for i,value in enumerate(driver.find_elements_by_xpath("//span[@class='ellipsis fleft fs12 lh16']")):
    if i%3 == 2:
        job_location.append(value.text)
    
df = pd.DataFrame({'job_title':job_titles,'job_company':job_company,'job_locations':job_location,'job_skills':job_skills})
df
```

```python
#9. Scrape the details most watched tv series of all time from imdb.com.Url = https://www.imdb.com/list/ls095964455/
#You have to find the following details:A) NameB) Year spanC) GenreD) Run timeE) RatingsF) Votes
```

```python
driver=webdriver.Edge("msedgedriver.exe")
    
driver.get("https://www.imdb.com/list/ls095964455/")

title_l,year_l,runtime_l,genre_l,votes_l,rating_l = [],[],[],[],[],[]

for title in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a"):
    title_l.append(title.text)
for year in driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']"):
    year_l.append(year.text)
for rating in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']"):
    rating_l.append(rating.text)
for runtime in driver.find_elements_by_xpath("//span[@class='runtime']"):
    runtime_l.append(runtime.text)
for genre in driver.find_elements_by_xpath("//span[@class='genre']"):
    genre_l.append(genre.text)
for votes in driver.find_elements_by_xpath("//span[@name='nv']"):
    votes_l.append(votes.text)


df = pd.DataFrame(list(zip(title_l,year_l,rating_l,genre_l,runtime_l,votes_l)),columns=['title','year_span','rating','genres','runtime','votes'])
df
```
